```python
#Import libraries
import pandas as pd
import numpy as np
```

```python
#Load the Health Data (Unemployment_data) CSV
health_data = pd.read_csv("Unemployment_data.csv")

health_data.head()
```

```python
#Drop the Indicator Code Column
health_data_clean = health_data.drop(columns=['Indicator Code'], inplace = True)
```

```python
#Move Each Year column to a single column
health_data_clean = pd.melt(health_data, id_vars=['Country Name', 'Country Code', 'Indicator Name'], var_name="Year", value_name="Value")
health_data_clean.head()
```

```python
#Filter only for the "Total Unemployment Row"
total_unemployment_data = health_data_clean.loc[health_data_clean['Indicator Name'] == 'Unemployment, total (% of total labor force)'] 

total_unemployment_data.head()
```

```python
#Write to csv
total_unemployment_data.to_csv('unemployment.csv',index=False)
```
