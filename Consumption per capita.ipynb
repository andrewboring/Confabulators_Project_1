```python
import pandas as pd
import numpy as np
```

```python
#load .csv 
coffee_data = pd.read_csv("data/project_data.csv")

coffee_data.head()
```

```python
#Drop Unneeded Columns
coffee_data.dropna(axis='columns', thresh=2, inplace = True)
```

```python
#Convert Coffee Consumed to Numeric

coffee_data['Coffee Consumed'] = coffee_data['Coffee Consumed'].astype('str')
coffee_data['Coffee Consumed'] = coffee_data['Coffee Consumed'].str.replace(',', '')
coffee_data['Coffee Consumed'] = pd.to_numeric(coffee_data['Coffee Consumed'])

#Calculate Coffee Consumed in kg
coffee_data['Coffee Consumed in KG'] = coffee_data['Coffee Consumed'] * 1000 * 60
coffee_data.head()
```

```python
#Calculate Coffee Consumed per capita in KG

coffee_data["Coffee Consumed per Capita in Kg"] = coffee_data['Coffee Consumed in KG'] / coffee_data['population']

coffee_data.head()
```

```python
coffee_data.to_csv("project_data_with_percapita_consumption.csv", index=False)
```
